```python
import xarray as xr
from pathlib import Path
```

```python
NWP_PATH = Path("/mnt/storage_b/data/ocf/solar_pv_nowcasting/nowcasting_dataset_pipeline/NWP/UK_Met_Office/UKV/native/2021/01/01/")
```

```python
filenames = list(NWP_PATH.glob("*Wholesale1.grib"))
```

```python
%%time
ds = xr.open_dataset(
    filenames[1],
    engine="cfgrib",

    # Wind direction and wind speed are the only variables which are at 
    # 10 meters above ground in the UKV GRIB files.
    #backend_kwargs={'filter_by_keys': {'level': 10}}
    backend_kwargs=dict(filter_by_keys={'stepType': 'instant'})
)
```

```python
%%time
ds2 = xr.open_mfdataset(
    filenames[:2],
    engine="cfgrib",
    combine="nested",
    concat_dim="time",

    # Wind direction and wind speed are the only variables which are at 
    # 10 meters above ground in the UKV GRIB files.
    #backend_kwargs={'filter_by_keys': {'level': 10}}
    backend_kwargs=dict(
        filter_by_keys={
            'stepType': 'accum', 
            'level': 10}
    )
)
```
